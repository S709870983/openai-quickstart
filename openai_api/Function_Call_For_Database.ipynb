```python
def get_all_tables(cursor):
    cursor.execute("SELECT table_name = t.name FROM sys.tables t INNER JOIN sys.schemas s ON t.schema_id = s.schema_id ORDER BY t.name;")
    tables = [row[0] for row in cursor.fetchall()]
    return tables

def get_columns_of_table(cursor, table_name):
    query = "SELECT column_name = c.name FROM sys.columns c INNER JOIN sys.tables t ON t.object_id = c.object_id WHERE t.name = N'{}';".format(table_name)
    cursor.execute(query)
    columns = [row[0] for row in cursor.fetchall()]
    return columns

def get_all_tables_and_columns(cursor):
    table_columns_dict = {}
    all_tables = get_all_tables(cursor)

    for table in all_tables:
        columns = get_columns_of_table(cursor, table)
        table_columns_dict[table] = columns

    return table_columns_dict
import pymssql

def query_sql():
    """
    通过pymssql连接SQL Server数据库并执行特定的SQL查询语句.
    主要是取出会议ID、会议日期、会议关联医院名称、speaker、会议主题、会议回放地址、会议海报
    """
    server = 'db.rds.wingwell.cloud'
    port = '20099'
    user = 'andy'
    password = 'my6eLQeojdiQ'
    database = 'wingwell'

    # 建立连接
    connect = pymssql.connect(server=server, port=port, user=user, password=password, database=database)
    cursor = connect.cursor()
    all_tables = get_all_tables(cursor)
    get_columns = get_columns_of_table(cursor, 'r_meeting')
    database_schema_dict = get_all_tables_and_columns(cursor)
    print(all_tables)
    print(get_columns)
    return database_schema_dict
```
